```python
def format_duration(seconds):
    minutes = seconds/60
    hours = minutes/60
    days = hours/24
    years = days/365
    if minutes>=1:
        seconds = seconds%60
    else:
        minutes =0
    if hours>=1 :
        minutes = minutes%60
    else:
        hours =0
    if days >=1 :
        hours = hours%24
    else:
        days =0
    if years >= 1: 
        days = days%365
    else:
        years = 0
    time = [int(years), int(days), int(hours), int(minutes), seconds]
    strng = ""
    x= []
    y = []
    for i in range(len(time)):
        if time[i] >=1:
            x = [time[a] for a in range(i,len(time))]
            y = len(list(filter(lambda x: x>=1, x)))
            if i ==0:
                if time[i] == 1:
                    strng+= str(time[i])+" year"
                else:
                    strng+= str(time[i])+" years"
            if i ==1:
                if time[i] == 1:
                    strng+= str(time[i])+" day"
                else:
                    strng+= str(time[i])+" days"
            if i ==2:
                if time[i] == 1:
                    strng+= str(time[i])+" hour"
                else:
                    strng+= str(time[i])+" hours"
            if i ==3:
                if time[i] == 1:
                    strng+= str(time[i])+" minute"
                else:
                    strng+= str(time[i])+" minutes"
            if i ==4:
                if time[i] == 1:
                    strng+= str(time[i])+" second"
                else:
                    strng+= str(time[i])+" seconds"
            if y >= 3: strng+=', '
            if y == 2: strng+= ' and '
    if strng == '':
        strng+= "now"
    return strng
```

```python
# Shorter
times = [("year", 365 * 24 * 60 * 60), 
         ("day", 24 * 60 * 60),
         ("hour", 60 * 60),
         ("minute", 60),
         ("second", 1)]

def format_duration(seconds):

    if not seconds:
        return "now"

    chunks = []
    for name, secs in times:
        qty = seconds // secs
        if qty:
            if qty > 1:
                name += "s"
            chunks.append(str(qty) + " " + name)

        seconds = seconds % secs

    return ', '.join(chunks[:-1]) + ' and ' + chunks[-1] if len(chunks) > 1 else chunks[0]
```
